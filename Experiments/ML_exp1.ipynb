```python
# Task 1: Import required libraries
# -------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots in Colab
%matplotlib inline
```

```python
# Task 2: Load Titanic Dataset
df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Successfully!\n")
```

```python
# Task 3: Display first 10 rows
print("First 10 Rows:")
print(df.head(10))
```

```python
# Task 4: Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail(5))
```

```python
# Task 5: Dataset Dimensions
print("\nDataset Shape:")
print(df.shape)
```

```python
# Task 6: Dataset Information
print("\nDataset Information:")
df.info()
```

```python
# Task 7: Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

print("\nAge Statistics")
print("Mean Age:", df["Age"].mean())
print("Minimum Age:", df["Age"].min())
print("Maximum Age:", df["Age"].max())

print("\nFare Statistics")
print("Mean Fare:", df["Fare"].mean())
print("Minimum Fare:", df["Fare"].min())
print("Maximum Fare:", df["Fare"].max())
```

```python
# Task 8: Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Task 9: Fill Missing Age Values
median_age = df["Age"].median()
df["Age"].fillna(median_age, inplace=True)

print("\nMissing Age Values after Filling:")
print(df["Age"].isnull().sum())
```

```python
# Task 10: Filter Surviving Passengers
survivors = df[df["Survived"] == 1]

print("\nNumber of Surviving Passengers:")
print(survivors.shape[0])

print("\nFirst 5 Surviving Passengers:")
print(survivors.head())
```

```python
# Task 11: Convert Fare Column to NumPy Array
fare_array = df["Fare"].to_numpy()

print("\nFare NumPy Array:")
print(fare_array)
```

```python
# Task 12: Maximum and Average Fare
print("\nMaximum Fare:", np.max(fare_array))
print("Average Fare:", np.mean(fare_array))
```

```python
# Phase 5: Data Visualization

sns.set(style="whitegrid")
```

```python
# Task 13: Histogram of Age
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Task 14: Bar Chart of Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df, palette="Set2")
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Task 15: Scatter Plot of Age vs Fare
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
